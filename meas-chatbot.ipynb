```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

# TinyLlama-1.1B-Chat: instruction-following, works great on Colab T4
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

# 4-bit quantization to fit in memory
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
)
model.config.use_cache = False

print("Model:", model_name)
print("Parameters:", sum(p.numel() for p in model.parameters()) / 1e6, "M")
print("✅ Model Loaded")
```

```python
from datasets import load_dataset

dataset = load_dataset("json", data_files="amr_shafek_dataset.json")

# TinyLlama uses ChatML format — wrap each sample properly
def tokenize_function(examples):
    texts = []
    for p, c in zip(examples["prompt"], examples["completion"]):
        # ChatML format that TinyLlama was trained on
        text = (
            "<|system|>\nYou are a helpful chatbot that answers questions about Amr Shafek, an AI Engineer.</s>\n"
            f"<|user|>\n{p}</s>\n"
            f"<|assistant|>\n{c}</s>"
        )
        texts.append(text)

    tokenized = tokenizer(
        texts,
        truncation=True,
        padding="max_length",
        max_length=256,
    )
    tokenized["labels"] = tokenized["input_ids"].copy()
    return tokenized

tokenized_dataset = dataset.map(tokenize_function, batched=True)
tokenized_dataset.set_format(
    type="torch",
    columns=["input_ids", "attention_mask", "labels"]
)

print("✅ Tokenization Done")
```

```python
from peft import LoraConfig, get_peft_model, TaskType, prepare_model_for_kbit_training

# Required before LoRA when using 4-bit quantization
model = prepare_model_for_kbit_training(model)

lora_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    r=8,
    lora_alpha=16,
    # TinyLlama uses LLaMA architecture
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    bias="none"
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
print("✅ LoRA Attached")
```

```python
import torch
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="./amr_chatbot_tinyllama",
    num_train_epochs=10,
    per_device_train_batch_size=1,
    gradient_accumulation_steps=4,
    fp16=True,
    dataloader_pin_memory=False,
    logging_steps=20,
    learning_rate=2e-4,
    save_total_limit=1,
    report_to="none",
    optim="paged_adamw_8bit",   # memory-efficient optimizer for 4-bit models
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
)

trainer.train()
print("✅ Training Done")
```

```python
# Save fine-tuned LoRA weights + tokenizer
model.save_pretrained("./amr_chatbot_tinyllama")
tokenizer.save_pretrained("./amr_chatbot_tinyllama")
print("✅ Model saved to ./amr_chatbot_tinyllama")

# Zip for easy download to your PC
import shutil
shutil.make_archive("amr_chatbot_tinyllama", "zip", "./amr_chatbot_tinyllama")
print("✅ Zipped as amr_chatbot_tinyllama.zip — download this to your PC")
```

```python
import torch
model.config.use_cache = True
model.eval()
def chat(user_input, max_new_tokens=100):
    prompt = (
        "<|system|>\nYou are a chatbot that only answers questions about Amr Shafek, "
        "an AI Engineer from Cairo, Egypt. Keep answers short and factual.</s>\n"
        f"<|user|>\n{user_input}</s>\n"
        "<|assistant|>\n"
    )

    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=512,
    ).to(model.device)

    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,          # greedy decoding — no randomness
            temperature=1.0,          # ignored when do_sample=False
            repetition_penalty=1.3,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
        )

    response = tokenizer.decode(
        output[0][inputs["input_ids"].shape[-1]:],
        skip_special_tokens=True
    ).strip()

    # Cut off at first newline to avoid rambling
    if "\n" in response:
        response = response.split("\n")[0].strip()

    return response if response else "I don\'t have information about that."


# Interactive chat loop
print("=" * 55)
print("     MEAS Chatbot (TinyLlama) — Ask about Amr Shafek")
print("     Type \'exit\' to quit")
print("=" * 55)

while True:
    user_input = input("\nYou: ").strip()
    if user_input.lower() in ["exit", "quit", "bye"]:
        print("Chatbot: Goodbye!")
        break
    if not user_input:
        continue
    print(f"Chatbot: {chat(user_input)}")
```
